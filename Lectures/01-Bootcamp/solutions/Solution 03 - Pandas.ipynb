1. Continuing "howlongwelive.csv", check whether which column has how many missing data

2. Fix all missing data using means

3. Perform a <code>groupby</code> country.  Which country has the lowest/highest life expectancy?

4. Perform a <code>groupby</code> status.  Is there any strong difference between developed and developing countries in their life expectancy?

```python
# 1. Continuing "howlongwelive.csv", check whether which column has how many missing data
import numpy as np
print("amount of missing data for each column:")
print(np.sum(data.isnull()))
```

```python
# 2. Fix all missing data using means
data.fillna(data.mean(),inplace=True)
print("No more missing data:")
print(np.sum(data.isnull()))
```

```python
# 3. Perform a <code>groupby</code> country.  Which country has the lowest/highest life expectancy?
grouped = data.groupby('Country')
series = grouped.mean()['Life expectancy ']
series.sort_values()
```

```python
# 4. Perform a <code>groupby</code> status.  Is there any strong difference between developed and developing countries in their life expectancy?
grouped = data.groupby("Status")
series = grouped.mean()['Life expectancy ']
print(series)
```

```python
# 5. Create another dataframe manually with 2 columns. First column is the ID column with same value as country column. Also add another column Noise_level, and populate with random values (whatever you like).  
countries = np.unique(data['Country'])
noises = np.random.uniform(50, 100, countries.shape)
noise_df = pd.DataFrame({'Country': countries, 'Noise_level': noises})
noise_df.head()
```

```python
# 6. Merge the two datasets together based on the ID column.
data = data.merge(noise_df, on="Country")
data.head()
```

```python
# 1. Load "resources/appl_1980_2014.csv" into df
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

apple = pd.read_csv("../data/appl_1980_2014.csv")
apple.head()
```

```python
# 2. Transform dta column to datetime type
apple.Date = pd.to_datetime(apple.Date)
```

```python
# 3. Then filter only with year not less than 1987
apple = apple[apple['Date'].dt.year > 1986]
```

```python
# 4. Sort the index in an ascending order (oldest date first) - use sort_index
apple = apple.set_index('Date')
apple.sort_index(ascending = True, inplace=True)
apple.head()
```

```python
# 5. We want to know how apple does on end of every month, thus get the mean value for each columns, resample at last buiness day of each month (i.e., BM)
apple_month = apple.resample('BM').mean()
apple_month
```

```python
# 6. BTW, how many months do we have in our data?
len(apple_month.index)
```

```python
# 7. Compare the differences between this year and last year High column, using shift(), where the differences is simply this_year - last_year.  Plot this graph, where x-axis is date, and y-axis is the gain/loss
apple_high = apple['High']
apple_high_last_year =apple_high.shift(365)
apple_high.plot(style=['g-'], label='this year')
apple_high_last_year.plot(style=['b:'], label='last year')
diff = apple_high - apple_high_last_year
diff.plot(style=["r--"], label='diff')
plt.legend()

#seems like we got lots of ups, but only in year 2002 that we see the dip
#likely because in 2000, apple overperform a bit
```

```python
# 8. Perform a rolling mean (moving average) of Close with window size of 100 days
apple_close = apple['Close']
apple_close.rolling(window=100).mean().plot(style='-', label="rolling")
apple_close.plot(style='--', figsize=(10, 7), label="normal")
plt.title("Whatever title you like")
plt.ylabel('Close Price', fontsize=14)
plt.legend()
```

```python
# 9. Load microsoft data using this code <code>microsoft = data.DataReader('MSFT', start='1987', end='2014', data_source='yahoo')</code>.  Compare Apple and Microsoft of their return rate of "close" price based on year 2000 onward, where the formula is simply <code>close price / close price [0]</code> where close price [0] is simply the first close price of year 2000
# pip3 install pandas_datareader 

from pandas_datareader import data

microsoft = data.DataReader('MSFT', start='1987', end='2014',
                      data_source='yahoo')

microsoft.head()

microsoft_close = microsoft['Close']

apple_close = apple_close['2000':]
microsoft_close = microsoft_close['2000':]

apple_return = apple_close.apply(lambda x: x / apple_close[0])
microsoft_return = microsoft_close.apply(lambda x: x / microsoft_close[0])

apple_return.plot(style=["g--"], figsize=(10, 7), label="Apple")
microsoft_return.plot(style=["r:"], label="MS")

plt.legend()
```
