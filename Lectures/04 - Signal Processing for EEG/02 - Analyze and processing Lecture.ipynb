#### Fourier Series

[quote](https://en.wikipedia.org/wiki/Fourier_series#:~:text=A%20Fourier%20series%20(%2F%CB%88f,be%20determined%20using%20harmonic%20analysis.)
> A Fourier series (/ˈfʊrieɪ, -iər/[1]) is a sum that represents a periodic function as a sum of sine and cosine waves. The frequency of each wave in the sum, or harmonic, is an integer multiple of the periodic function's fundamental frequency. Each harmonic's phase and amplitude can be determined using harmonic analysis. 

The key is for any signal that is periodic, the Fourier Series can represent it. 

Periodic signal is a signal that repeat after some period of time. A sum of periodic signal is periodic.

Period signal satisfy the next properties.
$$ x(t) == x(t + T_0) $$
$T_0$ is a period. 

The equation mean, there is signal *x* which is a function of time *t*. $T_0$ is some number. If this signal *x* repeats every 5 second ($T_0 = 5$) then x[0] == x[0 + 5], x[1] == x[1 + 5], ...

Every signal we have seen in the previous lecture/lab is periodic signal.

Thus, the sum of periodic signal is represented with a sum of sine and cosine at different frequencies, amplitudes, and phases.

The example of fourier series are all the wave in the previous lecture.

$$
x(t) = \frac{4}{\pi} \sum_{k=1}^{\infty} \frac{1}{2 k - 1} \sin \left( 2 \pi \left(2 k - 1 \right) t \right)
$$

Here is the square wave equation, we know right away that the square wave consists of odd harmonics. If we refactor the equation into a series of sum, it will look like this.

$$ 
x(t) = \frac{4}{\pi} \times ( \frac{\sin(2\pi t)}{1} + \frac{\sin(2\pi 3t)}{3} + \frac{\sin(2\pi 5t)}{5} + ... )
$$

For an easier to read formula, we encapulate the $2\pi t$ into $\omega$

$$ 
x(t) = \frac{4}{\pi} \times ( \frac{\sin(\omega)}{1} + \frac{\sin(3\omega)}{3} + \frac{\sin(5\omega)}{5} + ... )
$$

well, just to show you that it is a sum of sine wave. Thus, it is fourier transform.

##### Fourier Series General Form

Here is the general form of Fourier series.

$$
x(t) = a_0 + \sum_{n=1}^{\infty} a_n \cos(nt\frac{\pi}{L}) + \sum_{n=1}^{\infty} b_n \sin(nt\frac{\pi}{L})
$$

where *L* is half of the period and $a_0,a_n,b_n$ are the coefficient.

Fun fact:

Euler’s Equation is $ e^{i\theta} = \cos(\theta) + i\sin{\theta}$

and $ \cos{\theta} = \frac{e^{i\theta} + e^{-i\theta}}{2} $, $\sin{\theta} = \frac{-ie^{i\theta} + ie^{-i\theta}}{2}$

Thus the Fourier Series can be written as

$$
x(t) = a_0 + \sum_{n=-\infty}^{\infty} U_n e^{int\frac{\pi}{L}}
$$

Full derivation is [here](http://www.ee.ic.ac.uk/hp/staff/dmb/courses/E1Fourier/00300_ComplexFourier.pdf)

Here is the trick, all we have to find is the coefficient $a_0,a_n,b_n$ and here is how.

$$
a_0 = \frac{1}{2L} \int_{-L}^{L} x(t) dt
$$

$$
a_n = \frac{1}{L} \int_{-L}^{L} x(t) \cos(nt\frac{\pi}{L}) dt
$$

$$
b_n = \frac{1}{L} \int_{-L}^{L} x(t) \sin(nt\frac{\pi}{L}) dt
$$

$$
U_n = \frac{1}{L} \int_{-L}^{L} x(t) e^{-int\frac{\pi}{L}}  dt
$$

Don't get scared by the integration yet. My point is, to find the coefficient, we only consider 1 cycle. For mathematic nerd, continue to read [this](https://www.mathsisfun.com/calculus/fourier-series.html)


#### Fourier Transform

Not every signal is periodic. In fact, most of the signals are aperiodic. What is aperiodic signal? It is a signal that never repeats itself (yet).

$$ \forall{T_{0}} (x(t) != x(t + T_0)) $$

In human word, you can not find and number of $T_0$ that makes $x[t] == x[t + T_0]$ for all *t*. Or, there is no period/interval.

Therefore, instead of looking at the signal in a single period/interval $[-L,L]$ like the series, Fourier Transform look at the signal as the period is infinity. 

$$
F(x(t)) = X(f) = \int_{-\infty}^{\infty} x(t) e^{-i2\pi ft} dt
$$

We ditch the *L* which is half period which means there is no period in the equation.

$F()$ is a Forward Fourier Transform (FT) function. $x()$ is function of time representing a signal. $X()$ is a transformed signal which is now a function of frequency representing a signal in frequency domain.

We can reverse the operation using Inverse Fourier Transform (iFT) ($F^{-1}$)

$$
F^{-1}(X(f)) = x(t) = \int_{-\infty}^{\infty} X(f) e^{i2\pi ft} df
$$

#### Discrete Fourier Transform and Fast Fourier Transform

[How are the Fourier Series, Fourier Transform, DTFT, DFT, FFT, LT and ZT Related?](https://www.youtube.com/watch?v=2kMSLqAbLj4)

What we have learnt so far is the continous version of the function. For us, programmer, we don't have `countinuous` signal. In fact, we don't have a infinte long signal. When we are calculating, we have a finite reorded/sampled signal. This is where we have an discrete version of the function. Now, we take a look at Discrete Fourier Transform (DFT). And Fast Fourier Transform (FFT) is actually the faster version of DFT. 

Instead of integrating over the graph, we have a summation over the range of time.

$$
\text{FFT}(x[n]) = X[k] = \sum_{n=0}^{N-1} x[n] e^{-i2\pi k\frac{n}{N} }
$$

$$
\text{FFT}^{-1}(X[k]) = x[n] = \frac{1}{N} \sum_{n=0}^{N-1} X[k] e^{i2\pi k\frac{n}{N} }
$$

We change the notation from *t* to *n* which mean a sample of time and *f* to *k* which mean a sample frequency. *N* is the number of sample in calculation. 

Now that we have a functon, let's do this from scratch. It's fun.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal
import mne
plt.style.use('seaborn-whitegrid')
```

```python
# Create a known signal
fs = 1000
t = np.arange(start=0, stop=1+ 1/fs, step=1/fs)
x = np.sin(2 * np.pi * 10 * t + 0) # f = 10

plt.plot(t,x)
plt.show()
```

```python
N = len(x)
# let's find X[k] where k = 10
k = 10
X_10 = 0

for n in range(N):
    X_10 += x[n] * np.exp( -1j * 2 * np.pi * k * n / N )

print(X_10)

plt.scatter(k, abs(X_10))
plt.show()
```

The result we get is a complex number and the magnitude is 500?

Let's complete the other `k` and arrange them in the array of `X`.

```python
N = len(x)
# let's find X[k] 
X = []
for k in range(fs):
    # Use broadcasting instead of a loop
    X.append((x * np.exp( -1j * 2 * np.pi * k * np.arange(N) / N )).sum())

fig, ax = plt.subplots(nrows=2, figsize=(10,6))

ax[0].scatter(range(fs),np.abs(X))

ax[1].scatter(range(fs),np.abs(X))
ax[1].set_xlim(0,20)
ax[1].set_xlabel('k')
plt.show()
```

Ok, if we find the `k` in a range of `fs`, there are two peaks which are 10 and 990. Why? Temporal aliasing from last lecture. We can extend the range of `fs`, and we will get even more aliasing.

```python
N = len(x)
# let's find X[k] 
X = []
for k in range(-fs, fs):
    X.append((x * np.exp( -1j * 2 * np.pi * k * np.arange(N) / N )).sum())

plt.plot(range(-fs, fs),np.abs(X))
plt.xlabel('k')
plt.show()
```

You can see, when plot, both signals look the same. 

To the eye of DFT, it sees the same signal, both extended 10Hz and 20Hz.

Is this means we can not use DFT with signal longer than 1 second because the function will misinterpret the result? No.

The interpretation is up to us.

`k` is not a frequency. It will equal to the frequency if and only if `N` == sampling_rate

`k` is an index of an array. The content and meaning is up to us.

Thus, we can think of `k` as a smaller step of frequency. Instead of [0,1,2,3,...], we get [0,0.5,1,1.5,2,...]

Hence, when we increase the `N` either by extending the signal or padding the zero, we get a higher resolution of frequency range.

"What, we can have 0.5 Hz?" of course, we can. Oscillation frequency is not limited to an integer.

With this knowledge, we modify the DFT as follow.

```python
fs = 1000
t = np.arange(start=0, stop=2+ 1/fs, step=1/fs)
x = np.sin(2 * np.pi * 10 * t + 0)
# Try the padding version
# x[1000:] = 0
N = len(x)
freq_range = np.arange(start=0, stop=fs, step=fs/N)
X = []
for k in range(N):
    X.append((x * np.exp( -1j * 2 * np.pi * k * np.arange(N) / N )).sum())

fig, ax = plt.subplots(nrows=2, figsize=(10,5))
# The correct scale is N//2, but be careful, this N is the number of sample point with out padding the zero.
ax[0].plot(freq_range,np.abs(X)/(N//2))

ax[1].plot(freq_range,np.abs(X)/(N//2), marker='.')
ax[1].set_xlim(0,20)
ax[1].set_xlabel('frequency (Hz)')
plt.show()
```

Are you having fun?

Lets summarized what we have learned about DFT/FFT.

$$
\text{FFT}(x[n]) = X[k] = \sum_{n=0}^{N-1} x[n] e^{-i2\pi k\frac{n}{N} }
$$

*k* is the index of frequency array. The interpretation of what each *k* equal to which frequency, you need to map to the actual frequency range calculate separately.

*N* is the total number of sample in calculation.

Increasing the number of sample by either (1) extend the time range or (2) padding zero will yield a smaller step/ higher resolution of frequency range result.

`freq_range = np.arange(start=0, stop=fs, step=fs/N)`

The result *X* is a summation of data points after projected to the exponential function. To scale down with `N//2` where `N` is the number of data point before zero pading.

Let's finish this.

```python
def myDFT(x: np.ndarray, fs:int, padding:int = 0) -> tuple:
    """
    x: is a signal in numpy array. Right now onyl support (n_sample, ) shape
    fs: the sampling rate. This only be used for calculatin frequency_range
    padding: number of zero to pad to increase the size of `N`
    """
    scale_factor = len(x)//2
    # Padding
    shape = list(x.shape)
    shape[-1] = padding
    x = np.concatenate([x, np.zeros(shape)], axis=-1)
    # DFT
    N = len(x)
    k = np.arange(N).reshape(-1,1)
    n = np.arange(N).reshape(1,-1)
    X = np.abs((x * np.exp( -1j * 2 * np.pi * k * n / N )).sum(axis=-1)) / scale_factor
    # The DC component will be twice as much because of the mirror.
    X[0] = X[0] / 2
    freq_range = np.arange(start=0, stop=fs, step=fs/N)
    return (X, freq_range)

fs = 500
t = np.arange(start=0, stop=2 + 1/fs, step=1/fs)
x = 1 + np.sin(2 * np.pi * 10 * t + 0) + (0.5 * np.sin(2 * np.pi * 20 * t + 0)) + (0.8 * np.sin(2 * np.pi * 50 * t + 0))

# Play with the padding
X, freq_range = myDFT(x,fs,padding=0)

plt.plot(freq_range,X)
plt.xlabel('frequncy (Hz)')
plt.show()
```

Well, fortunately, NumPy has FFT function too. It supports flexible shape. This means you can do FFT on multiple signals at the same time.

```python
fs = 500
t = np.arange(start=0, stop=2 + 1/fs, step=1/fs)
x = 1 + np.sin(2 * np.pi * 10 * t + 0) + (0.5 * np.sin(2 * np.pi * 20 * t + 0)) + (0.8 * np.sin(2 * np.pi * 50 * t + 0))

X = np.fft.fft(x, axis=-1, n=None)
freq_range = np.fft.fftfreq(x.shape[-1], d=1/fs )
# You have to perform the absolute and the scale down yourself.
# Thus, the np.fft.fft only perform the sum of x[n] * expo[x] for us
plt.plot(freq_range,np.abs(X)/(len(x)/2))
plt.xlabel('frequncy (Hz)')
plt.show()
```

### 1.2 Short Time Fourier Transform (STFT)

As of now, The transformation we did is using the entire range of signal. It is fine when we want to have a brief look/analyze of the signal.

However, in the real world, we might interest in the change of frequency through time which the DFT/FFT does not provide.

In this matter, we use Short-Time-Fourier-Transform (STFT)

```python
fs = 1000
t = np.arange(start=0, stop=3 + 1/fs, step=1/fs)
f = np.linspace(start=3, stop=9, num=len(t))
x = np.sin(2 * np.pi * f * t + 0)

plt.plot(t,x)
plt.xlabel('time (second)')
plt.show()

X = np.fft.fft(x, axis=-1, n=None)
freq_range = np.fft.fftfreq(x.shape[-1], d=1/fs )
plt.plot(freq_range[:fs//2],np.abs(X[:fs//2])/(len(x)/2))
plt.xlabel('frequncy (Hz)')
plt.xlim(0,20)
plt.show()
```

The above example shows the non-stationary signal. This signal has a variable frequency. 
The result of DFT, not only incorrect, does not show the change of frequency shift though time.

Let's try to use STFT. Don't worry, we will not do from scratch. 

```python
fs = 1000
t = np.arange(start=0, stop=3 + 1/fs, step=1/fs)
f = np.linspace(start=3, stop=9, num=len(t))
x = 2*np.sin(2 * np.pi * f * t + 0) + 4*np.sin(2 * np.pi * (f+100) * t + 0) + 5*np.sin(2 * np.pi * (f*10) * t + 0)

# https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.stft.html
f_bin,t_bin,Zxx = scipy.signal.stft(x=x, fs=fs, nperseg=fs)

print(f_bin.shape, t_bin.shape, Zxx.shape)
c = plt.pcolormesh(t_bin, f_bin, np.abs(Zxx),cmap='plasma')
plt.colorbar(c)
plt.title('STFT Magnitude')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()

plt.specgram(x, Fs=fs,cmap='plasma')
plt.show()
```

Let's compare this to the `mne` STFT

```python
fs = 1000
t = np.arange(start=0, stop=3 + 1/fs, step=1/fs)
f = np.linspace(start=3, stop=9, num=len(t))
x = 2*np.sin(2 * np.pi * f * t + 0) + 4*np.sin(2 * np.pi * (f+100) * t + 0) + 5*np.sin(2 * np.pi * (f*10) * t + 0)

# https://mne.tools/stable/generated/mne.time_frequency.stft.html
X = mne.time_frequency.stft(x=x, wsize=fs, tstep=fs/2)

print(X.shape)
c = plt.pcolormesh(range(X.shape[-1]), range(X.shape[1]), np.abs(X.squeeze()), cmap='plasma')
plt.colorbar(c)
plt.title('STFT Magnitude')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

The STFT is actually is a FFT in smaller segments. The extra parameter we have to specify is `window size`/`nperseg` and `overlap`/`tstep`.

For the previous 2 STFT results, we use window size of `fs` and by default, the overlap is half of the window size (`fs/2`). 

Here I plot each time bin separately.

```python
fig, ax = plt.subplots(nrows=4, figsize=(10,10))
# freq_range = np.fft.fftfreq(x.shape[-1], d=1/fs )
# print(freq_range)
freq_range = np.arange(0,501,step=1)
for i in range(4):
    ax[i].text(x=200,y=0.2,s=f"{0.5*i-0.5} - {0.5*i+0.5} second")
    ax[i].plot(freq_range, np.abs(Zxx[:,i]))

ax[3].set_xlabel('Frequncy (Hz)')
plt.show()
```

Is this mean, we can segment the signal and perform the FFT ourselves?

## 2. Filters

Before we learn what is a filter, let's try to mimic the STFT result. We know, the STFT using filter named 'hanning'.


```python
fig, ax = plt.subplots(nrows=3, figsize=(10,10))
ax[0].plot(np.hanning(fs))
ax[0].text(x=0,y=0.8,s="Hanning window")

filtered_x = x[:1000] * np.hanning(fs)
ax[1].plot(x[:1000],label='Original',alpha=0.4)
ax[1].plot(filtered_x, label='Filtered')
ax[1].legend()

X_filtered = np.fft.fft(filtered_x, axis=-1, n=None)
freq_range = np.fft.fftfreq(1000, d=1/fs )
ax[2].plot(freq_range[:fs//2],np.abs(Zxx[:500,1]), label='STFT')
ax[2].plot(freq_range[:fs//2],np.abs(X[:fs//2])/(len(x[:1000])/2),label='FFT', alpha=0.5)
ax[2].plot(freq_range[:fs//2],np.abs(X_filtered[:fs//2])/(len(x[:1000])/2),label='FFT - filtered', alpha=0.3, c='red')
ax[2].legend()
ax[2].set_xlabel('frequncy (Hz)')
plt.show()
```

Aha!! The filtered version match exactly with the library one. Now, you know.

Now, we learn about filter.

Most often, this means removing some frequencies or frequency bands. However, filters do not exclusively act in the frequency domain; especially in the field of image processing many other targets for filtering exist. <u>My take</u>

You use a filter when you have a mixture of substance, and you want to remove some unwanted substance from the entire mixture.

Most common filter is air filter. In this case, the substance is air. Air consistent of a lot of things including PM2.5. We know that the PM2.5 particle is larger than the air, therefore we build a filter with a small hole whose size is smaller than PM2.5. When air pass through the filter, PM2.5 can not get pass the filter, so the air is cleaner.

In the same sense, signal contain a mixture frequency (if we look at the frequency-domain) or a mixture of multiple addition of signals (time-domain). 

Just like the air + PM 2.5, there are certain things you might not want in the signal. The things we don't want usually defined as `noise`. 

In summary, you use a filter when (1) Shaping the signal (Hanning widows for smoothen the signal), and (2) remove noise.

As you seen, the downside of Hanning window is the amplitude of the signal is reduced. This is the trade-off of the filter. Always remember, filter takes out something. There are other windows designed with different trade off.

Now, we take a look at filter that will remove certain frequencies. 

```python
fs = 1000
t = np.arange(start=0, stop=3 + 1/fs, step=1/fs)
f = np.linspace(start=3, stop=9, num=len(t))
# Notice the DC component of 5
x = 5 + 2*np.sin(2 * np.pi * f * t + 0) + 4*np.sin(2 * np.pi * (f+100) * t + 0) + 5*np.sin(2 * np.pi * (f*10) * t + 0)
# Add white noise
noise_white = 0.001 * np.random.normal(size=x.shape)
# Add power line
noise_power_line = 10 * np.sin(2 * np.pi * np.random.randint(low=48, high=52, size=x.shape) * t)

x_mix = x + noise_power_line + noise_white


plt.plot(t, x_mix)
plt.xlabel('time (second)')
plt.show()

f_bin,t_bin,Zxx = scipy.signal.stft(x=x_mix, fs=fs, nperseg=fs)

# Zxx.mean(axis=1).shape

plt.plot(f_bin, np.abs(Zxx.mean(axis=1)))
plt.title('Average of STFT along the frequency axis')
plt.xlabel('Frequency [Hz]')
plt.show()

```

### High/Low/Band pass/stop

The high/low/band pass/stop filter are like air filter. We look at the signal as a mixture of frequencies and only want a certain part of its. Maybe we don't want any signal higher than a 150 Hz. In this case, we use Low Pass/High stop filter with a threshold of 150 Hz. The name Low Pass comes from we let the lower than threshold frequency pass and the High stop comes from, you guess it, we filter out the frequency higher than the threshold. Therefore, Low pass and high stop is the same things. 

To me, I often see the name "High Pass" and "Low Pass" filter more often because they have the abbreviation of HPF and LPF respectively. 

Band pass/stop filter is, as the name suggest, let a range of frequency pass or stop. A band is defined by lower and higher threshold. The result of the band pass of [10,100] will be equal to the result of a HPF at 10 and LPF at 100 (given use the same filter type and parameter).

Just like the smooth filter, H/LPF has many types with different trade-off. The common one is `butterworth` filter.

```python
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.butter.html
# N: Order of filter
# Wn: Cut-off frequency
# btype: {'lowpass', 'highpass', 'bandpass', 'bandstop'}
# b: numerator coefficient 
# a: denominator coefficient
b, a = scipy.signal.butter(N=4, Wn=250, btype='lowpass', output='ba', fs=fs)
# b and a will be used to perform a filtered on the signal
# To visualize the filter, we look at its frequency response

# https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.freqz.html#scipy.signal.freqz
# w: The frequencies at which h was computed, in the same units as fs. By default, w is normalized to the range [0, pi) (radians/sample).
# h: The frequency response, as complex numbers.
w, h = scipy.signal.freqz(b=b, a=a, worN=fs//2)
plt.plot(fs//2*(w/np.pi),np.abs(h))
plt.axvline(250, color='black', linestyle='dashed', alpha=0.5) # cutoff frequency
plt.xlabel('Frequency (Hz)')
plt.ylabel('Response')
plt.show()
```

Let's look at the effect of different parameters

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(20,10))


for N in [1,2,4,8,16,32]:
    b, a = scipy.signal.butter(N=N, Wn=250, btype='lowpass', output='ba', fs=fs)
    w, h = scipy.signal.freqz(b=b, a=a, worN=fs//2)
    ax[0][0].plot(fs//2*(w/np.pi),np.abs(h), label=f"Order:{N}")

ax[0][0].axvline(250, color='black', linestyle='dashed', alpha=0.5) # cutoff frequency
ax[0][0].set_title('Different N')

for Wn in [10,50,100,200,300,400,450,490]:
    b, a = scipy.signal.butter(N=4, Wn=Wn, btype='lowpass', output='ba', fs=fs)
    w, h = scipy.signal.freqz(b=b, a=a, worN=fs//2)
    ax[0][1].plot(fs//2*(w/np.pi),np.abs(h), label=f"Wn:{Wn}")

ax[0][1].set_title('Different Wn')

for btype in ['lowpass', 'highpass']:
    b, a = scipy.signal.butter(N=4, Wn=250, btype=btype, output='ba', fs=fs)
    w, h = scipy.signal.freqz(b=b, a=a, worN=fs//2)
    ax[1][0].plot(fs//2*(w/np.pi),np.abs(h), label=f"btype:{btype}")

ax[1][0].set_title('Different between H/LPF')

for btype in ['bandpass', 'bandstop']:
    b, a = scipy.signal.butter(N=4, Wn=[100,400], btype=btype, output='ba', fs=fs)
    w, h = scipy.signal.freqz(b=b, a=a, worN=fs//2)
    ax[1][1].plot(fs//2*(w/np.pi),np.abs(h), label=f"btype:{btype}")

ax[1][1].set_title('Different between bandpass and bandstop')

for i in [0,1]:
    for j in [0,1]:
        ax[i][j].set_xlabel('Frequency (Hz)')
        ax[i][j].set_ylabel('Response')
        ax[i][j].legend()
plt.show()
```

Let's apply the filter on our signal.

We know from the STFT plot that our frequency of interest is below 150Hz. Therefore, we will use LPF at 150Hz. The order depends on how much aggressive you want from the filter. The common/default is 4.

```python
b, a = scipy.signal.butter(N=4, Wn=150, btype='lowpass', output='ba', fs=fs)
x_mix_lpf = scipy.signal.lfilter(b=b,a=a,x=x_mix)

fig, ax = plt.subplots(nrows = 2, figsize=(10,10))
ax[0].plot(t, x_mix, label='Original', alpha=0.7)
ax[0].plot(t, x_mix_lpf, label='LPF', alpha=0.7)
ax[0].set_xlabel('time (second)')
ax[0].legend()

f_bin,t_bin,Zxx_lpf = scipy.signal.stft(x=x_mix_lpf, fs=fs, nperseg=fs)

ax[1].plot(f_bin, np.abs(Zxx.mean(axis=1)), label='Origial', alpha=0.7)
ax[1].plot(f_bin, np.abs(Zxx_lpf.mean(axis=1)), label='LPF', alpha=0.7)
ax[1].set_title('Average of STFT along the frequency axis')
ax[1].set_xlabel('Frequency [Hz]')
ax[1].legend()
plt.show()
```

Know, I notice the DC component. In our signal, we could remove by using `signal - signal.mean()` because our signal is periodic. In the real world, using a HPF at 1 Hz would be a better idea but that depends. Some time we want to study a gentle change that exist between seconds which is a rhythm between [0-1]. For now, I use HPF.

```python
b, a = scipy.signal.butter(N=4, Wn=1, btype='highpass', output='ba', fs=fs)
x_mix_hpf = scipy.signal.lfilter(b=b,a=a,x=x_mix_lpf)

fig, ax = plt.subplots(nrows = 2, figsize=(10,10))
ax[0].plot(t, x_mix_lpf, label='Original', alpha=0.7)
ax[0].plot(t, x_mix_hpf, label='HPF', alpha=0.7)
ax[0].set_xlabel('time (second)')
ax[0].legend()

f_bin,t_bin,Zxx_hpf = scipy.signal.stft(x=x_mix_hpf, fs=fs, nperseg=fs)

ax[1].plot(f_bin, np.abs(Zxx_lpf.mean(axis=1)), label='Origial', alpha=0.7)
ax[1].plot(f_bin, np.abs(Zxx_hpf.mean(axis=1)), label='HPF', alpha=0.7)
ax[1].set_title('Average of STFT along the frequency axis')
ax[1].set_xlabel('Frequency [Hz]')
ax[1].legend()
plt.show()
```

Now, my signal is contaminated by a 50 Hz rhythms from power line. I will use a band stop filter at 49-51 Hz.

```python
b, a = scipy.signal.butter(N=4, Wn=[49,51], btype='bandstop', output='ba', fs=fs)
x_mix_bs = scipy.signal.lfilter(b=b,a=a,x=x_mix_hpf)

fig, ax = plt.subplots(nrows = 2, figsize=(10,10))
ax[0].plot(t, x_mix_hpf, label='Original', alpha=0.7)
ax[0].plot(t, x_mix_bs, label='HPF', alpha=0.7)
ax[0].set_xlabel('time (second)')
ax[0].legend()

f_bin,t_bin,Zxx_bs = scipy.signal.stft(x=x_mix_bs, fs=fs, nperseg=fs)

ax[1].plot(f_bin, np.abs(Zxx_hpf.mean(axis=1)), label='Origial', alpha=0.7)
ax[1].plot(f_bin, np.abs(Zxx_bs.mean(axis=1)), label='HPF', alpha=0.7)
ax[1].set_title('Average of STFT along the frequency axis')
ax[1].set_xlabel('Frequency [Hz]')
ax[1].legend()
plt.show()
```

Look like the band stop does do much. This is the time to use Notch Filter. The idea of Notch filter is the same as band stop but much much narrowband.

```python
fig, ax = plt.subplots(nrows=3, figsize=(10,16))

for Q in [1,10,20,30,40,50]:
    # https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.iirnotch.html
    # w0: the frequency to remove
    # Q: the quality. it works like order
    # fs: sampling rate
    b, a = scipy.signal.iirnotch(w0=50, Q=Q, fs=fs)
    w, h = scipy.signal.freqz(b=b, a=a, worN=fs//2)
    ax[0].plot(fs//2*(w/np.pi),np.abs(h), label=f'Q:{Q}')
ax[0].set_title('Notch filter with different Q value')
ax[0].set_xlabel('Frequency (Hz)')
ax[0].set_ylabel('Response')
ax[0].legend()

# I use 8 because it seems to remove the 50Hz the best in this signal.
b, a = scipy.signal.iirnotch(w0=50, Q=8, fs=fs)
x_mix_notch = scipy.signal.lfilter(b=b,a=a,x=x_mix_hpf)

ax[1].plot(t, x_mix_hpf, label='Original', alpha=0.7)
ax[1].plot(t, x_mix_notch, label='Notch', alpha=0.7)
ax[1].set_xlabel('time (second)')
ax[1].legend()

f_bin,t_bin,Zxx_notch = scipy.signal.stft(x=x_mix_notch, fs=fs, nperseg=fs)

ax[2].plot(f_bin, np.abs(Zxx_hpf.mean(axis=1)), label='Origial', alpha=0.7)
ax[2].plot(f_bin, np.abs(Zxx_notch.mean(axis=1)), label='Notch', alpha=0.7)
ax[2].set_title('Average of STFT along the frequency axis')
ax[2].set_xlabel('Frequency [Hz]')
ax[2].legend()

plt.show()
```
